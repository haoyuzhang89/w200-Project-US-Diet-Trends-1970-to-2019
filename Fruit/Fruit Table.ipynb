## No2 Grape Vs Year

```python
grape_yr = pd.read_csv('dataframe/A1_df.csv')  # only for the first time

grape_yr.set_index('Year',inplace = True)
grape_yr.index = fruit_yr.index.map(str)

grape_yr = grape_yr[[ 'Grapes, all', '  Fresh.1','  Canning.1', '  Juice.1', '  Dried (raisins)', '  Wine']]
grape_yr.columns = ['All', 'Fresh', 'Canning', 'Juice', 'Dried', 'Wine']


grape_yr['Fresh_prop'] = grape_yr.Fresh/grape_yr.All
grape_yr['Canning_prop'] = grape_yr.Canning/grape_yr.All
grape_yr['Juice_prop'] = grape_yr.Juice/grape_yr.All
grape_yr['Dried_prop'] = grape_yr.Dried/grape_yr.All
grape_yr['Wine_prop'] = grape_yr.Wine/grape_yr.All

grape_yr.to_csv('dataframe/A1_grape.csv')
```
