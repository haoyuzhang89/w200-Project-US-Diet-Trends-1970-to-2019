```python
### Grouping with an Index Sample

So far, we've always used a `Series` for our group key.  It's also possible to leverage dictionaries and functions when grouping.

For this demo, the data to group by will need to be in the `Index`.  Let's begin by setting the index to the `hotel_name` column
```

```python
Vegas.set_index('hotel_name').head()
```

```python
def f(name):
    if 'Circus' in name or 'Flamingo' in name:
        return 'preferred'
    else:
        return 'non-preferred'
```

```python
Vegas.set_index('hotel_name').groupby(f).mean()
```

### Total fruits , fresh, canning, freezing, juice, dried Vs year

```python
# loading from csv

total_yr = pd.read_csv('dataframe/A1_total_yr.csv')  

#total_yr.drop(columns=['Unnamed: 0'], inplace = True)

total_yr.set_index('Year',inplace = True)

total_yr.head(5)   ## Total all fruits, fresh, canning, freezing, juice Vs year
```

```python
total_yr.plot()
plt.legend(bbox_to_anchor=(1.01, 0.5), loc = "center left")
```

```python
plt.plot(total_yr.Fresh, label = 'Fresh')
plt.plot(total_yr.Juice, label = 'Juice')
plt.legend(bbox_to_anchor=(1.01, 0.5), loc = "center left")
```

```python
plt.plot(total_yr.Canning, label = 'Canning')
plt.plot(total_yr.Freezing, label = 'Freezing')
plt.plot(total_yr.Dried, label = 'Dried')
plt.legend(bbox_to_anchor=(1.01, 0.5), loc = "center left")
```

```python
total_yr["In refrigerator"] = total_yr.Fresh + total_yr.Freezing + total_yr.Juice
total_yr["Out refrigerator"] = total_yr.Canning + total_yr.Dried
```

```python
plt.plot(total_yr["In refrigerator"], label = "In refrigerator")
plt.plot(total_yr["Out refrigerator"], label = "Out refrigerator")
plt.legend(bbox_to_anchor=(1.01, 0.5), loc = "center left")
#plt.yscale('log')
```
