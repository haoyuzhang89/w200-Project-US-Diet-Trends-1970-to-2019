## Table G-18--Fresh oranges: Supply and utilization, 1980/81 to date								
	Supply				Utilization			
 Season 1	Utilized 							Per Capita
    	production 2	Imports	Total supply		Exports		Domestic	use  (Million pounds, pound for per capita use)


```python
df_g18 = pd.read_csv("Data_sourse All tables/Table_G18.csv")  # first time to run 

#df_a1.dropna(axis = 'rows', how ='all', inplace = True)
#df_a1.head()
#df_a1.set_index('Crop',inplace = True)

df_g18.dropna(axis = 'columns', how ='all', inplace = True)
df_g18 = df_g18[ 5: ]
df_g18.columns = ['Year', 'Production', 'Imports', 'Total supply', 'Exports',
       'Domestic use', 'Per capita use']

df_g18.set_index('Year',inplace = True)
for var in df_g18.columns:
    df_g18[var] = df_g18[var].map(lambda x : float(str(x).replace(',', '')))

df_g18.to_csv('dataframe/G18_df.csv')
#df_g18.columns
```

```python
df_g18 = pd.read_csv('dataframe/G18_df.csv')  # after the second time
df_g18.set_index('Year',inplace = True)
#fruit_yr.index = fruit_yr.index.map(str)
#print(fruit_yr.columns)

df_g18.head(5)  # whole dataframe table set up
```

```python
orange_in_ex = df_g18[['Production', 'Imports', 'Exports', 'Domestic use']]
orange_in_ex.plot()

plt.legend(bbox_to_anchor=(1.01, 0.5), loc = "center left")
plt.ylabel("Million pounds")
plt.xticks(rotation='vertical')

plt.title("Fresh oranges: Supply and utilization")
plt.yscale('log')

```
