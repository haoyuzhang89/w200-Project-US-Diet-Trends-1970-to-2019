## Total comsumption of meat fruit vege

```python
vege = pd.read_csv('../vege_total.csv')

vege = vege.T
#vege.columns = ['Year','Vegetable']

vege.columns = vege.iloc[0]
vege = vege[1:]

for var in vege.columns:
    vege[var] = vege[var].map(lambda x : float(str(x)))
    
vege.head(50)
                              
```

```python
fruit = pd.read_csv('../fruit_total.csv')
fruit = fruit.T
#fruit.columns = ['Year', 'Fruit']

fruit.columns = fruit.iloc[0]
fruit = fruit[1:]

for var in fruit.columns:
    fruit[var] = fruit[var].map(lambda x : float(str(x)))
    
#fruit.set_index('Year', inplace = True)

fruit.head(50)

fruit.columns
```

```python
meat = pd.read_csv('../meat_total.csv')

for var in meat.columns:
    meat[var] = meat[var].map(lambda x : float(str(x)))
    
meat.Year = meat.Year.map(lambda x : int(x))


meat.head(50)
```

```python
diet_part = pd.concat([ vege, fruit], axis=1)
```
