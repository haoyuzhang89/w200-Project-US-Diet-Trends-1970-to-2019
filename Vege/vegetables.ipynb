```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
df1 = pd.read_csv('Table_1.csv')

# split the type of vegetables (all, fresh, can and processed)
df1['Crop_Type'], df1['Type'] = df1['Crop'].str.split(',').str

# remove all the spaces
df1["Type"] = df1["Type"].str.strip()

df1 = df1.replace("---", np.nan)

df1.dtypes
```

```python
# Convert the columns of years to float
for col in df1.columns[1:51]:
    df1[col] = pd.to_numeric(df1[col], errors = "coerce")

df1.dtypes
```

```python
df1
```

# Summarize vegetables data and draw a total graph

```python
crop_all = df1.loc[df1["Type"] == "all"].T
# crop_all.loc["1970":"2019"].astype(float)
# crop_all.loc["Total"] = crop_all.sum()

crop_fresh = df1.loc[df1["Type"] == "fresh"].T
crop_can = df1.loc[df1["Type"] == "canning"].T
crop_freeze = df1.loc[df1["Type"] == "freezing"].T

vege_sum = pd.DataFrame()
vege_sum["All"] = crop_all.sum(axis = 1)[1:51]
vege_sum["Fresh"] = crop_fresh.sum(axis = 1)[1:51]
vege_sum["Canning"] = crop_can.sum(axis = 1)[1:51]
vege_sum["Freezing"] = crop_freeze.sum(axis = 1)[1:51]

# draw the all vegetables chart with level in bar and rate of change in line
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.bar(vege_sum.index, vege_sum["All"])
ax2.plot(vege_sum.index, vege_sum["All"].rolling(2).mean().pct_change() * 100, color = "red")

# set axis names and title
ax1.set_xlabel("Year")
ax1.set_ylabel("Pound per person")
ax2.set_ylabel("Rate of change")
ax1.set_title("Per Capita Availability of Selected Vegetables")
ax1.set_ylim(300, 500)
ax2.set_ylim(-5, 5)

# change the x-axis to show every three years
ax1.set_xticks(vege_sum.index[::3])
ax1.set_xticklabels(vege_sum.index[::3], rotation = 90)
ax2.yaxis.set_major_formatter(mtick.PercentFormatter())

plt.show()
```

# Absolute Level for three main categories

```python
fig, ax = plt.subplots(1, 2, figsize=(15,5))
x = vege_sum.index

# plot the three different types
ax[0].plot(x, vege_sum["Fresh"], color = "red", alpha = 0.5, label = "Fresh")
ax[0].plot(x, vege_sum["Canning"], color = "blue", alpha = 0.5, label = "Canning")
ax[0].plot(x, vege_sum["Freezing"], color = "green", alpha = 0.5, label = "Freezing")

# change the x-axis to show every three years
ax[0].set_xticks(vege_sum.index[::3])
ax[0].set_xticklabels(vege_sum.index[::3], rotation = 90)

# set axis names and title
ax[0].set_xlabel("Year")
ax[0].set_ylabel("Pound per person")
ax[0].set_title("Per Capita Availability of Selected Vegetables")
ax[0].legend(loc = "best")

# plot the three different types as a percentage of total
y1 = vege_sum["Fresh"] / vege_sum["All"] * 100
y2 = vege_sum["Canning"] / vege_sum["All"] * 100
y3 = vege_sum["Freezing"] / vege_sum["All"] * 100

ax[1].plot(x, y1, color = "red", alpha = 0.5, label = "Fresh")
ax[1].plot(x, y2, color = "blue", alpha = 0.5, label = "Canning")
ax[1].plot(x, y3, color = "green", alpha = 0.5, label = "Freezing")

# change the x-axis to show every three years
ax[1].set_xticks(vege_sum.index[::3])
ax[1].set_xticklabels(vege_sum.index[::3], rotation = 90)

# set axis names and title
ax[1].set_xlabel("Year")
ax[1].set_ylabel("Percentage")
ax[1].yaxis.set_major_formatter(mtick.PercentFormatter())
ax[1].set_title("Percentage as of All")
ax[1].legend(loc = "best")
```

### Rate of Change for the three main categories

```python
fig, ax = plt.subplots(1, 3, figsize=(15,3))

# plot the rate of change for the three types
x = vege_sum.index
y1 = vege_sum["Fresh"].rolling(2).mean().pct_change() * 100
y2 = vege_sum["Canning"].rolling(2).mean().pct_change() * 100
y3 = vege_sum["Freezing"].rolling(2).mean().pct_change() * 100

ax[0].plot(x, y1)
ax[0].set_title("Fresh")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("Rate of Change")
ax[0].set_xticks(x[::3])
ax[0].set_xticklabels(x[::3], rotation = 90)
ax[0].yaxis.set_major_formatter(mtick.PercentFormatter())

ax[1].plot(x, y2)
ax[1].set_title("Canning")
ax[1].set_xlabel("Year")
ax[1].set_xticks(x[::3])
ax[1].set_xticklabels(x[::3], rotation = 90)
ax[1].yaxis.set_major_formatter(mtick.PercentFormatter())

ax[2].plot(x, y3)
ax[2].set_title("Freezing")
ax[2].set_xlabel("Year")
ax[2].set_xticks(x[::3])
ax[2].set_xticklabels(x[::3], rotation = 90)
ax[2].yaxis.set_major_formatter(mtick.PercentFormatter())
```

### Detailed Analysis for each vegetable type
We are looking into each vegetable type. Based on per capita availability, we are dividing the vegetables into three categories based on data from 2019:
1. High availability: pound per capita > 20
2. Medium availability: pound per capita > 10
3. Low availability: pound per capita 5 - 10
4. Very low availability: pound per capita 0 - 5

Top 3 are: Potatoes, Tomatoes and Lettuce
Top 5 are: Onions and Sweet Corn

```python
cols = list(crop_all.iloc[51, :])
crop_all_val = crop_all.iloc[1:51, :]
crop_all_val.columns = cols

crop_all_val.plot()
plt.legend(bbox_to_anchor=(1, 0.5), loc = "center left")

# filter out the high, medium and low availability vege types
crop_all_val.T.sort_values(by = ['2019'], ascending = False)
crop_all_high = crop_all_val.T[crop_all_val.T["2019"] > 20].T
crop_all_med = crop_all_val.T[(crop_all_val.T["2019"] < 20) & (crop_all_val.T["2019"] > 10)].T
crop_all_low = crop_all_val.T[(crop_all_val.T["2019"] < 10) & (crop_all_val.T["2019"] > 5)].T
crop_all_vlow = crop_all_val.T[(crop_all_val.T["2019"] < 5) & (crop_all_val.T["2019"] > 0)].T

# draw the four charts
fig, ax = plt.subplots(2, 2, figsize=(15,10))

crop_all_high.plot(ax = ax[0, 0])
ax[0,0].set_title("High Availability")
ax[0,0].legend(bbox_to_anchor = (0.5, 0.9), frameon=False, loc='lower center', ncol = 4)
ax[0,0].set_ylim(0, crop_all_high.max().max() + 20)

crop_all_med.plot(ax = ax[0,1])
ax[0,1].set_title("Medium Availability")
ax[0,1].legend(bbox_to_anchor = (0.5, 0.9), frameon=False, loc='lower center', ncol = 4)
ax[0,1].set_ylim(0, crop_all_med.max().max() + 5)

crop_all_low.plot(ax = ax[1,0])
ax[1,0].set_title("Low Availability")
ax[1,0].legend(bbox_to_anchor = (0.5, 0.82), frameon=False, loc='lower center', ncol = 3)
ax[1,0].set_ylim(0, crop_all_low.max().max() + 3)

crop_all_vlow.plot(ax = ax[1,1])
ax[1,1].set_title("Very Low Availability")
ax[1,1].legend(bbox_to_anchor = (0.5, 0.82), frameon=False, loc='lower center', ncol = 3)
ax[1,1].set_ylim(0, crop_all_vlow.max().max() + 1)

plt.show()
```

### Percentage change for top 5
Here we will look at the percentage change for the top 5 vege types.

Top 3 are: Potatoes, Tomatoes and Lettuce
Top 5 are: Onions and Sweet Corn

```python
crop_all_pct = pd.DataFrame()
for col in crop_all_val.columns:
    col_name = col + "_pct"
    crop_all_pct[col_name] = crop_all_val[col] / vege_sum["All"]
    
crop_all_pct_top5 = crop_all_pct[["Potatoes_pct", "Tomatoes_pct", "Lettuce_pct", "Onions_pct", "Sweet potatoes_pct"]]
# crop_all_pct_top5.head()

crop_all_pct_top5.plot()
plt.legend(bbox_to_anchor=(1, 0.5), loc = "center left")
plt.ylabel("Percentage")
plt.title("Top 5 Vegetable Types: 1970-2019")
plt.show()
```
