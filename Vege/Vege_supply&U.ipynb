```python
# Module & Package  - Run here

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns

%matplotlib inline
plt.style.use('default')
```

Table 34--U.S. fresh potatoes: supply, availability, & price, farm weight.1								



```python
df_t34 = pd.read_csv("Data source_All Tables/Table_34.csv")  # Potatoes

#df_a1.dropna(axis = 'rows', how ='all', inplace = True)
#df_a1.head()
#df_a1.set_index('Crop',inplace = True)

df_t34 .dropna(axis = 'columns', how ='all', inplace = True)
df_t34  = df_t34[ 3: ]
df_t34 .columns = ['Year', 'Production','Imports',' Total Supply', 'Exports', 'Domestic Availability', 'Per Capita Availability', 'Current Dollars', 'Constant 2012 Dollars']


df_t34 .set_index('Year',inplace = True)

df_t34.head(10)
```

```python
potato_in_ex = df_t34[['Production','Imports', 'Exports', 'Domestic Availability']]

for var in potato_in_ex.columns:
    potato_in_ex[var] = potato_in_ex[var].map(lambda x : float(x))

potato_in_ex.plot()

plt.legend(bbox_to_anchor=(1.01, 0.5), loc = "center left")
plt.ylabel("Million pounds")
plt.xticks(rotation='vertical')

plt.title("Fresh potatoes: Supply and availability")
```

```python
matrix = np.triu(potato_in_ex.corr())
sns.heatmap(potato_in_ex.corr(),annot=True, mask=matrix,vmin=-1, vmax=1, center= 0, cmap= 'coolwarm')
plt.xticks(rotation=0)
```
